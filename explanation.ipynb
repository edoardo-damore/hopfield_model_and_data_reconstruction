The notebook "MonteCarlo" aims to simulate a Hopfield network by the means of Monte Carlo methods using Metropolis algorithm.
Here are some explanations on how should it work:

- The first part creates a series of 1d pattern (i.e. vectors), which can be retrieved by the network using Metropolis algorithm
- The second one does the same, but with 2d patterns (i.e. matrices)

## List of 1d functions

- **create_patterns_1d(number of spins, number of patterns)**: return an array with 1d patterns 

- **compute_weights_1d(array of patterns)**: create the weight matrix for storing the patterns in memory

- **compute_energy_1d(a single pattern, weight matrix)**: given a single pattern within the generated set, we can compute its energy

- **pattern_corruption_1d(a single pattern, probability of a spin flip)**: given as input the probability of doing a random flip in the selected pattern, this function flips a number of spins accordingly to the given probability. returns the corrupted pattern.

- **metropolis_1d(corrupted pattern, weight matrix, Temperature, number of iterations)**: application of the metropolis algorithm to the corrupted pattern for retrieval process.

- **check_correctness(original pattern, retrieved pattern)**: checks the accuracy of the retrieval process. Returns a float between 0.0 and 1.0 

## List of 2d functions

- **create_patterns_2d(rows,columns,N_patterns)**: create an array of N_patterns elements composed by matrices "rows x columns" of spins.

### The other functions have the same name of the 1d case, except that a "2d" instead of a "1d". The only function that remains the same is "check_correctness".

Changing the temperature values, one should see different behaviours of the system:
- ) For low termperatures (approx. 0), the system should behave like the deterministic case
- ) For high termperatures, random flips should appear (due to thermal noise).
